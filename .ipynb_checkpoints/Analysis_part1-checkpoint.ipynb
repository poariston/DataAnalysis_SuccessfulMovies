#### The question is: What parameters are correlated with a movie with a low profit_ratio and a high profit_ratio?

### Feature engineering
#### make a categorical data low/medium/high based on profit_ratio <1  1-2 >2

```python
profit_lmh = []
for profit_ratio in df4['profit_ratio']:
    if profit_ratio <1:
        profit_lmh.append('low')
    elif  profit_ratio <2:
        profit_lmh.append('medium')
    else:
        profit_lmh.append('high')
print(len(profit_lmh))
df4['profit_lmh'] = profit_lmh
df4
```
