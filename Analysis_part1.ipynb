### Data cleaning

#### Make a list of each unique genre.

```python
l= list(df5.genres.unique())
print (l)
# split each elem accoring to ","
l2=''
for elem in l:
    l2=l2 + ',' + elem
list_genres_withduplicates = l2.split(',')
list_genres_withduplicates=list_genres_withduplicates[1:]
list_genres_withduplicates
```

```python
unique_genres=[]
for elem in list_genres_withduplicates:
    if elem not in unique_genres:
        unique_genres.append(elem)
unique_genres 
```

```python
# create columns for each genre to flag movie genre: Fill in with True if movie belongs to the genre 

for genre in unique_genres:
    l=[]
    for i in df5.index:
        if genre in df5['genres'][i]:
            l.append(True)
        else:
            l.append (False)
            
df5[genre]=l

df5.iloc[9, 14:]
```

#### Now I am able to analyse profitability according to genre 

```python
df5.shape
```

```python
dico_low={}
dico_high={}
for genre in  unique_genres:
    dico_low [genre]=len(df5[(df5[genre] == True) & (df5['profit_lmh'] == 'low') ])
    dico_high[genre]=len(df5[(df5[genre] == True) & (df5['profit_lmh'] == 'high') ])
```

```python
dico_low
```

```python
dico_high
```

```python
dico_low.keys() == dico_high.keys() 
```

```python
dflow=pd.DataFrame.from_dict(dico_low,orient='index',columns=['number_of_movies_lowp'])
dfhigh=pd.DataFrame.from_dict(dico_high,orient='index',columns=['number_of_movies_highp'])
```

```python
# dflow['profit_flag'] = 1
# dfhigh['profit_flag']=2


# df6=dflow.append(dfhigh,ignore_index=True)

df6=dflow.join(dfhigh,how='inner')
df6
```

```python
df6['total_by_genre']=df6['number_of_movies_lowp']+df6['number_of_movies_highp']
df6['fraction_highp_by_genre']=df6['number_of_movies_highp']/df6['total_by_genre']
df6['fraction_lowp_by_genre'] =df6['number_of_movies_lowp'] /df6['total_by_genre']
df6.reset_index(inplace=True)
df6=df6.rename(columns={'index':'genre'})
df6['ratio_num_highlow']= df6 ['fraction_highp_by_genre'] / df6['fraction_lowp_by_genre']
df6=df6.sort_values(by='ratio_num_highlow',ascending=True)
df6
```

#### Visualization of profitability according to genre
